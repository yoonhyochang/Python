#### while 문
- 반복횟수가 명확하지 않을 때

```python
# 파이썬 최고 ! 13번 출력해보기
num = 0
while num <= 13: 
    print(f'파이썬 최고! {num}번 출력')
    num += 1
```

```python
num = 1 
while True:
    print('무한루프')
    num += 1
    if num >3:
        break
```

```python
while True:
    num1 = int(input('첫 번째 정수 입력 >>'))
    num2 = int(input('두 번째 정수 입력 >>'))
    sum1 = num1 + num2
    if sum1 != 0:
        print(f'두 정수의 합 : {sum1}')
    elif sum1 == 0:
        print('프로그램이 종료 되었습니다.')
        break
```

```python
nw = int(input('현재 몸무게'))
gw = int(input('목표 몸무게'))
num = 1
while True:
    cw = int(input(f'{num}주차 감량 몸무게 : '))
    num+=1
    nw-= cw
    if gw >= nw:
        print (f'{gw}kg 달성! 축하합니다!!')
    
```

#### for 문
- 반복 횟수가 정해져 있을 때

```python
lunch = ['샌드위치', '백반','돈까스']
lunch
```

```python
for food in lunch :
    print (food)
```

```python
# 문자열 반복
hi = '안녕하세요'
for i in hi:
    print(i, end='') #출력문의 끝 속성 : end = ''
    # print 속성 : 기본값은 개행해서 출력
```

```python
score_list = [90, 45, 70, 60, 55]
num = 0
for a in score_list:
    num+=1
    if a >= 60:
        print(f'{num}번 학생은 합격입니다')
    else:
        print(f'{num}번 학생은 불합격입니다')
```

#####  for : range()
- 필요한 만큼의 숫자를 만들어내는 유용한 기능
- range(시작할 숫자, 종료할 숫자, 증감량)

```python
for i in range(0,len(lunch),1): 
    #시작인데스가 0과 같다면 생략가능, 증감값(defoult : 1)--> 생략 가능
    print(i,lunch[i])
```

```python
#range
score_list = [90, 45, 70, 60, 55]
for i in range(0,len(score_list),1):
    if score_list[i] >= 60:
        print(f'{i+1}번 학생은 합격입니다.')
    else:
        print(f'{i+1}번 학생은 불합격입니다.')
    
```

```python
for i in range(1,25,1):
    print(i, end= " ")
```

```python
# 1~100 사이의 숫자중 3의 배수의 합을 구하기
sum1 = 0
for i in range(0,100,3):
    sum1 += i          
print(sum1)  
```
