```python
# return을 사용하여 함수를 정의할 때
def divisor(num):
    for i in range(1,num+1):
        if num % i ==0:
            return i # 함수를 수행하면서 값 하나만을 반환하고 종료
        
```

```python
divisor(10)
```

```python
# return을 사용하여 함수를 정의하고 싶을 때
#--> 나중에 결과값을 사용하기 위해 별도의 변수에 저장
def divisor(num):
    list1 = []
    for i in range(1, num+1):
        if num % i ==0:
            list1.append(i)
    return list1
divisor(10)
```

#### return키워드 정리
- 반환값은 언제나 하나(ex. num_add, num_sub = nuber_m(num1,num2)
    ==>결과값은 (num_add, num_sub) : 패킹된 형태로 반환해줌
- 함수 안에서 return은 1번 실행
- 함수 안에서 return을 만나면 함수가 끝난다.

```python
# 더하고, 빼는 기능을 하는 함수 정의
def add_sub(n1,n2) :
    return n1+ n2, n1-n2
```

```python
#함수 호출
add_sub(5, 2)
```

```python
f1, s2 = add_sub(5,2)
print(f1,s2)
```

##### 가변 매개변수(*)
- 함수 호출 시 몇개의 인수가 전달될 지 알 수 없다면 사용자가 직접 매개변수의 개수를 정할 수 있도록 선언

```python
def add(*args):
    return sum(args)
    print(args)
# 가변 매개변수를 받아서 모든 숫자를 더한 결과를 반환하는 함수
# sum() 내장함수 이용
```

```python
add(1,2,3,4,5)
```

#### 가변 매개변수(**)
- 딕셔너리 형태로 함수 내부를 처리하고 싶을 때

```python
def print_map(**kwargs):
    print(kwargs)
    for k, v in kwargs.items():
        print(k,":",v)
```

```python
print_map(하나 = 1, 둘 = 2)
```
