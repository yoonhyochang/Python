#### 비교 연산자

```python
# True, false 로 결과가 출력되는 연산자
#>, >=, <=, == (같다),  !=(다르다) 
```

```python
a = 3
b = 7

print(a > b) # False
print(a <= b) #True
print(a == b) # False
print(a != b) #True
```

##### 논리 연산자

```python
#not : 논리를 뒤집어버리는 연산자 True -> False , false -> True
# and, or
```

```python
#not
a,b = 5,10
not a < b # True -> False
```

```python
# and, or
# and : 두 개의 논리를 연결하여 확인, 두 논리가 모두 참일 경우 : True
# or : 한 개의 논리만 참일 경우에도 True
3 > 5 and 10 == 10 #False
3 > 5 or 10 == 10 #True
```

#### 삼항 연산자
- True 실행문장 if 조건식 else False 실행문장

```python
# 60점 이상이면 합격, 미만이면 불합격 출력해보기
```

```python
성적 = 70
60 >= 성적
'합격'if 성적 >=60 else '불합격'
'불합격'if 성적 < 60 else '합격'
```
