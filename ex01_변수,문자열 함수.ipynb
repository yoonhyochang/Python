#### 문자열 함수

```python
#day 문자열에서 '햇볕 쨍쨍'인덱스 번호 확인하기
#find 함수 활용
a = day.find('햇볕 쨍쨍')
day[a:a+5]
```

```python
# indaex 함수
day.index('일')
```

```python
#find,index 함수 차이비교 :문장안에소 없는 문자열 찾을때
day.find('e')# 없을경우에 -1반환, 에러가 발생하지 않는 상황으로 더 많이 사용된다.
day.index('e') # 문자를 찾을 수 없다면 에러 발생
```

```python
# len() : 문자열 개수 반환(길이 반환)
len(day) # 공백,!,?,. 모두 포함한 길이
```

```python
# replace('찾을 문자열','변경할 문자열'):특정 문자열 찾아서 변경하는 함수
# day : 했볕 쨍쨍 -->비
day = day.replace('햇볕 쨍쨍','비')
day
```
