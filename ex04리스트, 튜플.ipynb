#### list 리스트
 - 파이썬 자료구조 중 하나
 - 순서가 있고, 수정 가능한 자료구조
 - 대괄호([])로 작성되고, 리스트 내부의 값은 콤마(,) 로구분
 - 하나의 리스트에 다양한 자료형을 포함
 - 추가,수정, 삭제가 가능

##### 리스트 선언

```python
# 리스트명 = [요소1, 요소2, 요소3...]
a = [] # 빈 리스트
b = [1, 2, 3] # 숫자요소를 담은 리스트
c = ['김시우','강로운','오경진','박유정','박서연']# 문자열 요소를 담은 리스트
d = [1,2,'이현수','임소완','윤효창']# 숫자, 문자열 요소를 담은 리스트
e = [1,2, '이현수',['임소완','윤효창']] # 리스트 내부에 리스트를 포함한 리스트

# [[]]: 이중리스트
# [] : 단일리스트
```

```python
print(a)
print(b)
print(c)
print(d)
print(e)
```

##### 리스트에 데이터 접근 : 인덱싱 , 슬라이싱

```python
# 인덱싱 : 리스트명[인덱스]
list1 = [1, 3, 5, 7, 9]
# 인덱스 번호 0에 접근
print(list1[0])
# 7요소에 접근
print(list1[3])
```

```python
temp = ['a', 'b', 'c']
list2 = [1, 2, 3, temp]

print(list2[3][1]) # 이중리스트에서 원하는 요소에 접근하는 인덱싱
```

```python
# len() : 길이를 알려주는 함수
len(list2)
```

 - 슬라이싱

```python
# 리스트명[시작인덱스 : 끝인덱스]
# 시작인덱스 <= 인덱스 번호 < 끝인덱스
list3 = ['김시우', '강로운', '오경진', '박유정', '박서연']
#'강로운','박유정' 데이터 접근 출력
list3[1:3] # 출력결과도 리스트로 반환
```

```python
#'김시우'~'오경진' 접근해보기
list3[:3] # 시작 인덱스 번호 생략 가능
```

```python
# '오경진'~ '박서연' 접근해보기
list3[2:]# 끝 인덱스 번호 생략 가능
```

```python
#인덱스 번호 0,2,4 접근해보기
list3[::2]
list3[1::2]# 인덱스 1번부터 끝까지 2개씩 띄어서 가져오기
```

```python
# 이중리스트를 만들고 데이터 접근해보기
# 이름, 나이, 혈액형 정보기반의 이중 리스트
list4 = [['김세미',25,'O'],['김보원',26,'O'],['이탁연',39,'AB'],
       ['임소완',25,'O'],['조연경',27,'A']]
list4
```

```python
# 이탁연의 나이, 혈액형 접근해서 출력해보기
print(f'탁연씨의 나이는 {list4[2][1]}이고, 혈액형은 {list4[2][2]}입니다.')
```

```python
# 임소완씨의 정보와 조연경씨의 정보를 리스트로 가져오기
# 나이와 혈액형을 출력해보기
print(list4[3:])
print(f'임소완씨 나이는{list4[3][1]} 혈액형은 {list4[3][2]}이고 연경씨 나이는{list4[4][1]} 혈액형은{list4[4][2]} 입니다')
```

#### 튜플 tuple
- 파이썬의 자료구조 형태 중 하나
- 순서가 있는 집합(인덱싱, 슬라이싱 가능)
- 소괄호 ()로 작성되어지고, 튜플 내부의 값은 콤마','로 구분
- 하나의 튜플에는 다양한 자료형을 포함
- 데이터의 추가 , 수정, 삭제가 불가능하다.


##### 튜플 선언


```python
a = () # 빈 튜플
b = (1, 2, 3) # 숫자형 데이터를 담은 튜플
c = ('박서연','박유정', '김다희')# 문자열 데이터를 담은 튜플
d = (1,2,'이도여', '이경영') # 숫자, 문자열 데이터를 담은 튜플
e = (("인공지능",'김시우',26),("A반",'이현수',26))# 이중 튜플

print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
type(e)
```

##### 튜플 데이터 접근 : 인덱싱, 슬라이싱

```python
# 튜플명[인덱스번호]
# 김시우의 나이 접근해보기
e[0][2]
```

```python
# 이름과 나이 가져오기(슬라이싱)
# 시우씨 이름과 나이 접근해보기
e[0][1:]
# 현수씨 이름과 나이 접근해보기
e[1][1:]
```

```python
list1 = ['딸기','바나나','복숭아','체리','사과']
a = input('검색할 문자를 입력하세요>> ')

if a in list1 != false:
    n = list1.index(a)
    print(f'{a}는 리스트에 들어 있지 않습니다.')   
else:
    print(f'{a}는 리스트{n}번째 인덱스에 들어 있습니다. ')
```
