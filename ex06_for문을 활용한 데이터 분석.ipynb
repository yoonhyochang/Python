#### 파이썬에서 파일을 읽고 쓰기
- 파이썬 내장함수 open()사용
    - f.open() ~ f.close()
- with문 사용
    -with open() as f:
     f.write()
    

#### pickle
- 파이썬에서 리스트, 튜플 등의 자료구조 형태를 유지하면서 파일로 저장, 불러올 때 사용하는 모듈
- 규칙이 있는 데이터의 형식을 유지하면서 데이터를 저장
- 바이너리 형태로 저장 (저장용량이 다른 파일 구조에비해 작다)
- 불러올때 : rb(read binary), 저장할때 : wb(write binary)

#### 네이버 영화 리뷰 데이터 분석하기

```python
import pickle
```

```python
#with문을 사용하여 파일을 불러오기
with open('movie_review.pkl','rb') as f:
    movie = pickle.load(f)
movie[0:5] #데이터 확인
# 0 : 영화에 대한평가
# 1 : 영화에 대한 감성평가 - 0 : 부정, 1 : 긍정
```

```python
#전체 리뷰데이터 수(개수) 확인하기
# len() 사용해서 리뷰의 개수 구하기
len(movie)
```

```python
#전체 리뷰 길의 평균값을 구해보자
#0번 인덱스의 리뷰의 길이
a = 0 # 리뷰 전체 길이를 담아줄 ㄴ변수

for i in movie:
    a += len(i[0])
    b = a / len(movie)
print(f'리뷰의 평균 길이{b}입니다')
```

```python
# 긍정 / 부정  리뷰의 개수를 구해보자
# 긍정 / 부정 리뷰 데이터에 접근(인덱싱)
a = 0
b = 0
for i in movie:
    if i[1] == 1:
        a+=1
    else:
        b+=1
print(f'긍정 리뷰의 수 : {a}개')
print(f'부정 리뷰의 수 : {b}개')
```

#### 멜론차트 Top100 활용해서 데이터를 분석해보고 나만의 차트를 만들기

```python
# 멜론차트 데이터를 불러와서 데이터 확인해보기
# with 문 사용해서 데이터 불러오기
with open('music_100.pkl','rb') as f:
    music = pickle.load(f)
    print(music)
```
