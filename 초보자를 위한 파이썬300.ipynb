- min, max 함수
- 선언한 자료구조(리스트, 튜플)에서 가장 큰 값, 가장 작은 값을 찾아주는 함수

```python
list2 = [4,5,2,1,99,15,2,7,27]
min(list2)
```

```python
max(list2)
max_num = list2[0]

for i in list2:
    if max_num < i:
        max_num = i
print(max_num)
```
